I film che nel titolo hanno la parola &quot;the&quot;

```python
result = db.movies.find({"title" : {"$regex" : "the"}},{"title" : 1})
result
df1 = pd.DataFrame(list(result))
df1
```

I film americani degli anni 20

```python
result = db.movies.find({'$and': [{"countries": "USA"},{"year": {'$gt' : 1919 }},{"year":{'$lt' : 1930 }}]},{"title" : 1,"year":1})
result
df1 = pd.DataFrame(list(result))
df1
```

```python
result = db.movies.find({'$and': [{"countries": "USA"},{"year": {'$gt' : 1929 }},{"year":{'$lt' : 1950 }}]},{"title" : 1,"year":1})
result
df1 = pd.DataFrame(list(result))
df1
```

Tutti i film tranne quelli americani

```python
result = db.movies.find({"countries": { "$nin": ["USA"]}},{"title" : 1,"countries":1})
result
df1 = pd.DataFrame(list(result))
df1
```

I film che hanno un rating superiore a 4

```python
result = db.movies.find({"imdb.rating" : {'$gt' : 4 }},{"title" : 1,"imdb.rating":1})
result
df1 = pd.DataFrame(list(result))
df1
```

Il numero di film per ogni lingua. Ordinare in ordine decrescente a
partire della lingua più presente

```python
result = db.movies.aggregate([ 
                                   {"$group" : {"_id": "$languages", "total" : {"$sum" : 1}}} , 
                                   { "$sort": { "total": -1 } } 
                                   ])
result
df1 = pd.DataFrame(list(result))
df1
```

Le prime 3 lingue in cui sono stati girati i film

```python
result = db.movies.aggregate([ 
                                   {"$group" : {"_id": "$languages", "total" : {"$sum" : 1}}} , 
                                   { "$sort": { "total": -1 } } ,
                                   { "$limit" : 3 }
                                   ])
result
df1 = pd.DataFrame(list(result))
df1
```

Le lingue in cui sono stati girati meno di 50 film

```python
result = db.movies.aggregate([ 
                                   {"$group" : {"_id": "$languages", "total" : {"$sum" : 1}}} , 
                                   { "$match": { "total": { "$lt": 50 } }},
                                   { "$sort": { "total": -1 } } 
                                   ])
result
df1 = pd.DataFrame(list(result))
df1
```

La durata media dei film italiani

```python
result = db.movies.aggregate([     {"$unwind" : "$languages"},
                                   { "$match": { "languages": "Italian"} },
                                   {"$group" : {"_id" : "null", "averageRunTime": { "$avg": "$runtime" }}}
                              
                                   ])
result
df1 = pd.DataFrame(list(result))
df1
```

La durata media dei film delle varie nazioni

```python
result = db.movies.aggregate([     {"$unwind" : "$languages"},
                                   {"$group" : {"_id" : "$languages", "averageRunTime": { "$avg": "$runtime" }}}
                              
                                   ])
result
df1 = pd.DataFrame(list(result))
df1
```

Il film che è dura di più

```python
result = db.movies.aggregate([     {"$group":{
           "_id": "$runtime",
           "maxRuntime": { "$max": "$runtime" }}},
           { "$sort": { "maxRuntime": -1 }},
           { "$limit" : 1 }])
result
df1 = pd.DataFrame(list(result))
df1
```
